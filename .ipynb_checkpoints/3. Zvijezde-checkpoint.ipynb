```python
import numpy as np
import matplotlib.pyplot as plt

def P1(x):
    return 1-x**2
def P2(x):
    return 1-(24./5.)*x**2+(28./5.)*x**3-(9./5.)*x**4
def M1(x):
    return x**3
def M2(x):
    return 4.*x**3-3.*x**4
def T1(x):
    return 1-x**2
def T2(x):
    return (1-(24./5.)*x**2+(28./5.)*x**3-(9./5.)*x**4)/(1-x)
# Example data
x = np.arange(0,1.2, .02)
m1=M1(x)
m2=M2(x)
p1= P1(x)
p2= P2(x)
t1=T1(x)
t2=T2(x)


f = plt.figure(figsize=(20,8))
ax=plt.subplot(131)
ax.plot(x, m1, 'k--', label='$\\rho(r)=\\rho$')
ax.plot(x, m2, 'k-', label='$\\rho(r)=\\rho_C\left(1-r/R\\right)$')
legend = ax.legend(loc='upper right', shadow=True)
ax.set_xlabel('$r/R$',size=20)
ax.set_ylabel('$M(r)/M$',size=20)
ax.set_xlim(0,1)
ax.set_ylim(0,1.2)
for label in legend.get_texts():
    label.set_fontsize('large')
ax=plt.subplot(132)
ax.plot(x, p1, 'k--', label='$\\rho(r)=\\rho$')
ax.plot(x, p2, 'k-', label='$\\rho(r)=\\rho_C\left(1-r/R\\right)$')
legend = ax.legend(loc='upper right', shadow=True)
ax.set_xlabel('$r/R$',size=20)
ax.set_ylabel('$P(r)/P_C$',size=20)
ax.set_xlim(0,1)
ax.set_ylim(0,1.2)
for label in legend.get_texts():
    label.set_fontsize('large')
ax=plt.subplot(133)
ax.plot(x, t1, 'k--', label='$\\rho(r)=\\rho$')
ax.plot(x, t2, 'k-', label='$\\rho(r)=\\rho_C\left(1-r/R\\right)$')
legend = ax.legend(loc='upper right', shadow=True)
ax.set_xlabel('$r/R$',size=20)
ax.set_ylabel('$T(r)/T_C$',size=20)
ax.set_xlim(0,1)
ax.set_ylim(0,1.2)
for label in legend.get_texts():
    label.set_fontsize('large')
#plt.savefig('ua1_modeli.png',bbox_inches='tight')
f
```

```python

def a(x):
    return 1-(24/5)*x**2+(28/5)*x**3-(9/5)*x**4
def ader(x):
    return (x/5)*(-48+84*x-36*x**2)
def M2(x):
    return abs(ader(x)/(ader(x)+a(x)/(1-x)))
# Example data
x = np.linspace(0,1.2)
m1=np.array([1. for y in x])
stabilnost=2.5*m1
m2=M2(x)



f=plt.figure()
ax=plt.subplot(111)
ax.plot(x, m1, 'k--', label='$\\rho(r)=\\rho$')
ax.plot(x, m2, 'k-', label='$\\rho(r)=\\rho_C\left(1-r/R\\right)$')
ax.plot(x, stabilnost, 'r-')
legend = ax.legend(loc='upper right', shadow=True)
ax.set_xlabel('$r/R$',size=20)
ax.set_ylabel('$|d\ln P/d\ln T|$',size=20)
ax.set_xlim(0,1)
ax.set_ylim(0,5)
for label in legend.get_texts():
    label.set_fontsize('small')
plt.savefig('ua2_konvekcija.png',bbox_inches='tight')
f
```
